## The Data
The dataset analysed in this project is the Labour Force Survey (LFS). This is a survey of private households in the UK, providing information on the labour market. Topics covered in the survey include economic activity; occupation; employment; health; education and demographic information. The LFS was conducted biannually from 1973 to 1983. It was then carried out annually until 1991 and has since been published in quarterly waves, demonstrating it to be a reliable dataset. The survey is intended to be representative of the UK population and its samples are considerably large, containing approximately 25,000 households per wave. It has a rotational sample design, whereby a household is selected and retained for five consecutive quarters, with 20% of the sample being replaced each quarter. However, this project analyses only the Q2 wave (April – June 2024); this was the most recent version available at the commencement of the project. The wide range of demographic and economic variables in the LFS make it an appropriate source to use in the analysis of the gender pay gap.

Data for the LFS is generated through telephone and face-to-face interviews. Participants include all those in the household who are aged 16 and over. Both core and non-core questions are asked, and demographic information is carried over to the next wave. The non-core questions vary with each wave but are updated regularly to reflect the current situation and priorities in relation to the UK labour force.

The dataset uses 348.9+ MB of memory and is organised in columns and rows, containing 825 variables and 55430 cases. Variables are both categorical and continuous but are mostly coded with numerical value labels. The dataset was acquired from the UK Data Service (UKDS), which is a trusted and highly reliable repository of datasets. Accessing the LFS requires registration to the UKDS and accepting the End User Licence. Each wave of the LFS has a rigorous Performance and Quality Monitoring Report, in which the quality of each survey is described. Due to the nature of the LFS, missingness is expected as data for certain variables only exist in particular waves. Therefore, the original sample size of 55430 cases is expected to reduce considerably, but still be of sufficient size for analysis in this project. The methodological process and the length of time which the LFS has been conducted for proves the survey to be high-quality.

### Creating Subsets of Boxplot Outliers

This dataset shows outliers in the variables Gross Pay per Week) and Usual Hours Worked per Week. The following code allows us to identify a subset of the specific points that are outliers in a column according to the boxplot.

```python
def outliers(data,column):
  '''Input the column as a string to make a subset of all outliers in a column \naccording to the outliers displayed in the boxplots. '''
  import pandas as pd
  import numpy as np
  d = data[column]
  iqr = np.percentile(d,75) - np.percentile(d,25)
  lb = np.percentile(d,25) - 1.5*iqr
  ub = np.percentile(d,75) + 1.5*iqr
  subset1 = data.loc[data[column] > ub]
  subset2 = data.loc[data[column] < lb]
  outliers = pd.concat([subset1,subset2])
  return outliers
```

## Objective 3: Gender, Disability and Ethnicity Pay Gap Analysis


* GPG =  Mean (average) and median % difference in hourly pay between male and female workers.
* DPG = Mean (average) and median % difference in hourly pay between disabled and non disabaled workers.
* EPG = Mean (average) and median % difference in hourly pay between workers of ethnic miniority background compared to the white ethnicity.
* GPG by Ethnicity = Mean (average) and median % difference in hourly pay between male and female workers of the same ethnicity.









```python
#Adding an hourly pay column to the dataframe to calculate GPG
lfs['Hourly_pay'] = (lfs['Gross Weekly Pay (£)']/lfs['Usual Weekly Hours Worked'])
```

```python
def meanGPG(df, gender, pay):
  '''Function that calculates the mean GPG using different parameters '''
  mean_male_pay = df[df[gender] == 'Male'][pay].mean()
  mean_female_pay = df[df[gender] =='Female'][pay].mean()
  mean_GPG = ((mean_male_pay - mean_female_pay)/mean_male_pay)* 100
  return (mean_GPG)
mean_gpg = meanGPG(df=lfs, gender='Sex', pay='Hourly_pay')
print(f"Mean gender pay gap: {mean_gpg:.2f}%")


def medianGPG(df, gender, pay ):
  ''' Function that calculates the median GPG using different parametres '''
  median_male_pay = df[df[gender] == 'Male'][pay].median()
  median_female_pay = df[df[gender] =='Female'][pay].median()
  median_GPG = ((median_male_pay - median_female_pay)/median_male_pay)* 100
  return (median_GPG)
median_gpg = medianGPG(df=lfs, gender='Sex', pay='Hourly_pay')
print(f"Median gender pay gap: {median_gpg:.2f}%")
```

### Disablity Pay Gap

```python
def DPG(df, disablity_status, pay):
  '''Function that calculates the mean disability Pay Gap using disabality status and hourly pay'''
  disabled_pay = df[df[disablity_status] == 'Disabled'][pay].mean()
  non_disabled_pay = df[df[disablity_status] =='Not Disabled'][pay].mean()
  mean_DPG = ((non_disabled_pay - disabled_pay)/non_disabled_pay)* 100
  return (mean_DPG)

mean_dpg = DPG(lfs, disablity_status='Disability', pay='Hourly_pay')

print(f"Mean Disability pay gap: {mean_dpg:.2f}%")

def medianDPG(df, disablity_status, pay):
  '''Function that calculates the median disability Pay Gap using disabality status and hourly pay'''
  disabled_pay = df[df[disablity_status] == 'Disabled'][pay].median()
  non_disabled_pay = df[df[disablity_status] =='Not Disabled'][pay].median()
  median_DPG = ((non_disabled_pay - disabled_pay)/non_disabled_pay)* 100
  return (median_DPG)

#Using the above function
median_dpg = medianDPG(lfs, disablity_status='Disability', pay='Hourly_pay')

print(f"Median Disability pay gap: {median_dpg:.2f}%")
```

### Ethnicity Pay Gap Vs White

```python
def ethPG(df, ethnicity, pay):
  '''Function that calculates the mean ethnicity pay gap compared to the White ethnicity'''
  mean_white_pay = df[df[ethnicity] == 'White'][pay].mean()
  ethnicity_pay_gaps = {}
  for group in lfs['Ethnicity'].unique():
    if group != 'White':
      eth_pg = df[df[ethnicity] == group][pay].mean()
      mean_EPG = ((mean_white_pay - eth_pg)/mean_white_pay)* 100
      ethnicity_pay_gaps[group] = mean_EPG
  return (ethnicity_pay_gaps)

#Using the above function
mean_epg = ethPG(lfs, ethnicity ='Ethnicity', pay='Hourly_pay')

def medianethPG(df, ethnicity, pay):
  '''Function that calculates the median ethnicity pay gap compared to the White ethnicity'''
  median_white_pay = df[df[ethnicity] == 'White'][pay].median()
  median_ethnicity_pay_gaps = {}
  for group in lfs['Ethnicity'].unique():
    if group != 'White':
      med_eth_pg = df[df[ethnicity] == group][pay].median()
      median_EPG = ((median_white_pay - med_eth_pg)/median_white_pay)* 100
      median_ethnicity_pay_gaps[group] = median_EPG
  return (median_ethnicity_pay_gaps)

#Using the above function
median_epg = medianethPG(lfs, ethnicity ='Ethnicity', pay='Hourly_pay')

#Formatting the final outputs
ethnicity_pg_df = pd.DataFrame({
    'Ethnicity': mean_epg.keys(),
    'Mean_Pay_Gap': [f"{gap:.2f}%" for gap in mean_epg.values()],
    'Median_Pay_Gap': [f"{median_epg[group]:.2f}%" for group in mean_epg.keys()]
})
ethnicity_pg_df = ethnicity_pg_df.set_index('Ethnicity')

ethnicity_pg_df = ethnicity_pg_df.rename(columns={
    'Mean_Pay_Gap': 'EPG (Mean)',
    'Median_Pay_Gap': 'EPG (Median)'
})
```

## 4a. Clustering

An exploratory k-means clustering will be carried out on the data, with k = 2. The aim is to explore whether the natural clusters formed within the data align with the categories 'Male' and 'Female'. If they do, this could suggest that the 'Sex' variable has a strong influence on the variability in within the data and could indicate a notable pay gap associated with gender. k-means has been chosen as opposed to other clustering algorithms because of the ability to set k arbitrarily.

Since the aim of the clustering is purely exploratory and not predictive, the data will not be split into training and test sets.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(lfs_c)

# kmeans using full dataset (no train/test split)
kmeans = KMeans(n_clusters=2, random_state=42)

cl_kmeans = kmeans.fit_predict(scaled_data)

clust_data = lfs_c.copy().reset_index()
clust_data['Cluster'] = cl_kmeans
clust_data.head()

# create a confusion table of sex and cluster by grouping the data
table = clust_data.groupby(['SEX', 'Cluster'])[['index']].count().unstack()

# convert into percentages to show the proportions of male vs female in each cluster
table_percent = table.divide(table.sum(axis = 0), axis = 1).mul(100).round(1)

def kmeans_confusion():
  ''' This function takes produces the heatmap for the confusion table showing the proportion of male and female
  observations in each cluster
  '''
  plt.figure(figsize=(4, 2.5))
  sns.heatmap(table_percent, annot=True, cmap = 'PRGn', fmt='.1f', cbar_kws={'label': 'Percentage (%)'})
  plt.title('Cluster Distribution by Sex (Percentages)')
  plt.ylabel('Sex')
  plt.xlabel('Cluster')
  plt.show()
```

```python
plt.figure(figsize = (4,3))

filtered_data = clust_data[clust_data['GRSSWK'] <= 2000]

sns.stripplot(
    data=filtered_data,
    x='SEX',
    y='GRSSWK',
    hue='Cluster',
    jitter=0.2,
    palette='PRGn',
    size=1.5
)

plt.title('Weekly Pay vs. Sex (Colored by Cluster)')
plt.xlabel('Sex')
plt.ylabel('Weekly Pay')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

### Silhouette score
The silhouette score is a measure of fit for cluster configurations. Scores close to 1 indicate good fit and well-separated cluster, while scores close to 0 indicate poor model fit and overlapping clusters. Negative values suggest incorrect cluster assignments.

```python
silhouette_score(scaled_data, cl_kmeans).round(3)
```

### Elbow plot
An elbow plot is a method of deciding the optimal number of clusters for a dataset. The 'elbow' where the inertia reduces most significantly appears to be at k = 4

```python
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(lfs_c)
    inertia.append(kmeans.inertia_)

# Plot inertia
plt.figure(figsize = (4,3))
plt.plot(range(1, 11), inertia)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Plot')
plt.show()
```

Loop over different values of k to see what the silhouette score would be for different numbers of clusters.

```python
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled_data)
    score = silhouette_score(scaled_data, labels)
    print(f'k={k}, Silhouette Score={score:.3f}')
```

The first objective we had was to clean the data so that it was ready for analysis of descriptive statistics and to build predictive and classification models with.

We started by examining the original dataset, which had 825 columns and 55,430 rows/participants. We then created a subset of the variables of interest, reducing the dataset to 10 columns.

Next, we assessed the dataset for missing values. Histograms of a few selected variables revealed many missing values, marked as negative values in the visualisations. Since missing data would skew the results in a linear model, we removed all rows containing any missing values, reducing the dataset to 10 columns and 7,385 rows.

The cleaned subset was uploaded to GitHub for easier access. When we reloaded the data, we found an unnecessary column containing the original row indices, which we removed.

We then renamed the columns with more interpretable labels to facilitate analysis and visualisation. Additionally, we created dictionaries to recode numerical values representing categorical data into strings, allowing us to produce bar plots for categorical variables in our descriptive statistics.

Next  we successfully removed 237 participants who were under 21 years old. This was because participants who were under 21 years old may skew the gross weekly pay data as there are different minimum wage restrictions for people under 21 years old compared to those over 21 years old. This resulted in the dataset consisting of 7148 people.

Using descriptive statistics, boxplots, and histograms, we identified outliers to be investigated. Outliers were calculated using the upper bound of 1.5x(interquartile range) above the 3rd quartile and lower bound of 1.5x(interquartile range) below the 1st quartile.

‘Usual Weekly Hours Worked’ had 916 outliers split fairly evenly between both ends of the distribution. The histogram and boxplot showed a symmetrical shape with a narrow central section and skewed tails. Since ‘Usual Weekly Hours Worked’ was an exploratory variable, we kept the outliers to maintain variation in the dataset, ensuring it reflected the population without impacting the primary analysis of pay and gender.

For ‘Gross Weekly Pay’, the boxplot showed severe positive skew due to outliers. We found 379 outliers at the higher end and none at the lower end. We excluded the top 1% highest earners to reduce the skew and make the distribution more normal for the linear model. This reduced the dataset from 24,411 participants to 7,070 participants.

## Objective 3

### Explanation of Results
In the project analysis, we utilised two metrics to measure the extent of gender, disability and ethnicity pay gap, within the labour force survey data. This is because the mean values are often affected by outliers (of which there are many outliers present in the dataset), whilst the median pay gap values, provide a more balanced view of the average pay gap negating the effect of outliers which skew the dataset.

The mean and median gender pay gaps respectively: 16.38 %: 16.05%, indicates a consistent pay gap in hourly earnings across all genders and ethnicities within the data. This could be due to many factors such as differences in working hours, roles and industries and of course underlying systematic bias and inequalities.

For full time vs part time workers, the mean and median GPG respectively is: 11.45%, 11.56% for full time workers and 11.33%, -2.74% for part time workers. The GPG has small variance between the mean and median for full time workers, suggesting a significant pay inequality favoured towards men of 11-12%. This could be due to variation in job roles, positions or opportunities between men and women who are full time workers. More interestingly there is a negative median pay gap for part time workers, which suggests women make marginally more than men in this area. This could be due to a few factors such as a higher number of men taking up lower paying roles amongst part time workers, in addition to more women occupying higher paying part time roles, due to childcare responsibilities, maternity leave etc. Further, the significant difference between the mean and median GPG values suggests a skewed distribution of earnings i.e. few highly paid part time male workers may be skewing the mean value.

Observing the gender distribution of pay quartiles show that like the initial GPG metrics, women are disproportionately in the lower paid quartiles Q1 and Q2 and men take over Q4 the highest pay quartile, contributing significantly to the overall gender pay gap. The mean and median disability pay gap: 12.58% and 12.11% respectively remains very consistent, indicating minimal effect from outliers. Overall, the disability pay gap is seen to be significant at 12% suggesting workers with disabilities earn up to 12% less than those without. This could be due to issues such as workplace bias and discrimination, lack of accessibility in roles for those with disabilities to progress equally, as well as lack of exposure to highly paid roles.

The ethnicity pay gap analysis shows that the Bangladeshi, Black/African/Caribbean/Black and Mixed/Multiple Ethnic Groups. These ethnicities have the highest median gap gaps at 28.98%, 14.01% and 11.21% respectively. These extremely large pay disparities indicate the presence of systemic issues. Some of the remaining ethnic groups show negative pay gaps i.e. Indian/ Other ethnic group, this could indicate the other socioeconomic factors to be at play, or perhaps overrepresentation in highly paid jobs. The Chinese ethnic group shows the lowest level of EPG, indicating near equality in pay in the median value but slightly higher mean value due to outliers.

The gender pay gap by ethnicity suggests there is minimal uniformity across different ethnicities. For example, Indian and White groups show the most significant GPG inequalities, at 16.31% and 17.35% respectively, although Bangladeshi women on average earn more than Bangladeshi men with a negative pay gap of -17.38%. Whilst Chinese ethnic group have a significant median GPG indicating pay inequalities from high earners, and the Black/African/Caribbean/Black British group has a low median GPG implying much smaller differences in pay distribution for Black men and women, though their mean GPG is higher than the median. Overall, this data suggest that both Ethnicity , gender and disability intersect to produce unique pay dynamics, where some ethnic groups experience more significant pay disparities than others.

### Explanation of Results

The confusion table above shows that Cluster 0 is composed of approximately 50% male and 50% female observations, whereas Cluster 1 consists of 22% males and 78% females. This shows that the clusters do not explicity align with the 'male' and 'female' categories,indicating that sex alone is not an influential variable in how the data naturally clusters. The silhouette score for the configuration with k = 2 clusters is 0.246, indicating poor fit. This is apparent in the scatterplot above, where we can see that the green and orange points (denoting the two clusters) are overlapping and the clusters are difficult to distinguish.

Looping over different values of k and evaluating the silhouette score for each configuration shows that the best clustering was in fact the original one with k = 2, with the others being around 0.1. The fact that there is no value of k that results in well-separated clusters could indicate that the data is not well-suited for clustering, or that further investigation into the data is required.The fact that no cluster configuration seems to align with categories of single variables suggests that no one variable is enough to understand differences in pay. This prompts investigation of all variables' effects using a linear model.

The second part of objective 4 set out to model pay against other sociodemographic variables. As detailed in the analysis, the results of the final model found sex, age, hours worked, occupation type, disability status and educational qualification to be statistically significant factors in predicted pay. Future modelling of pay should include a wider range of variables, because the variation of the models was only moderately predicted, as indicated by the relatively low adjusted R-Squared results. We successfully assessed the dataset for outliers and missing values. We removed all missing values which improved the accuracy of future analyses, particularly the OLS regression. #### Limitations
Some limitations include only selecting 10 variables to explore in our analysis. This reflects why our model did not have a considerably high adjusted R squared value. Including more variables may explain more of the variation in pay differences in the population. For example, someone’s pay could be influenced by their health, number of dependents, access to transport, or whether they are in receipt of government financial help.
Not all of the outliers were able to be excluded due to concern of compromising validity of the study by removing too many participants. Only one type of clustering, k-means, was used on the data, with the pre-processing only involving a standard scaler.

#### Future
